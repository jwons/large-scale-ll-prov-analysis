# Scripts that might have encountered rdtLite errors

These scripts had 'abnormal' details in their provenance including zero procedure nodes or zero data nodes in the presence of multiple procedure nodes. 

```python
import pandas as pd
prov_df = pd.read_csv("../output/raw_prov_table.csv")
```

```python
missing_nodes_df = prov_df[(prov_df.num_of_proc_nodes == 0) | (prov_df.num_of_data_nodes == 0)][["doi", "script_name", "num_of_proc_nodes", "num_of_data_nodes"]]
missing_nodes_df.to_csv("../output/missing_nodes.csv", index=False)
missing_nodes_df
```
